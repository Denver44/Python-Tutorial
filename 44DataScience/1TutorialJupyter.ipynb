```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 25.0, 0.07)
s = 1 + np.sin(5*np.pi*t)
plt.plot(t, s)

plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('About as simple as it gets, folks')
plt.grid(True)
plt.savefig("test.png")
plt.show()
```
