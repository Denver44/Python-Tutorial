```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.plot([1,2,3],[20,40,60]) # first x axis then Y axis
plt.title('EV CAR TREND')
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.show()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x1=[1,2,3]
y1=[10,20,30]
x2=[4,5,6]
y2=[40,50,600]
plt.plot(x1,y1,label="first",linewidth="4")
plt.plot(x2,y2,label="second")
plt.legend()
plt.show()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x1=[1,2,3]
y1=[10,20,30]
x2=[4,5,6]
y2=[40,50,600]
plt.scatter(x1,y1,label="first")
plt.scatter(x2,y2,label="second")
plt.legend()
plt.show()
```
